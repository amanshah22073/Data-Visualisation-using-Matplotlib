```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
columns = ['DATE', 'TAG', 'POSTS']
df = pd.read_csv("QueryResults.csv" , names=columns , header = 0)
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df
```

```python
pivot_df = df.pivot(columns = "TAG" , index = "DATE" , values = "POSTS")
```

```python
pivot_df
```

```python
pivot_df.shape
```

```python
pivot_df.tail()
```

```python
pivot_df.info()
```

```python
pivot_df.fillna(0 , inplace = True)
```

```python
pivot_df
```

```python
pivot_df.isna().values.any()
```

```python
plt.figure(figsize = (15,8))
plt.xlabel("tags"  ,fontsize = 15)
plt.ylabel("Dates" , fontsize = 15)
for column in pivot_df:
    plt.plot(pivot_df[column], label = column )
plt.legend(fontsize = 14)    
```

```python
average_df = pivot_df.rolling(window = 12).mean()
```

```python
average_df
```

```python
plt.figure(figsize = (15,8))
plt.xlabel("tags"  ,fontsize = 15)
plt.ylabel("Dates" , fontsize = 15)
for column in average_df:
    plt.plot(average_df[column], label = column )
plt.legend(fontsize = 14)  
```
